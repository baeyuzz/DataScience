```
import numpy as np
import scipy
import cv2 as cv
from google.colab.patches import cv2_imshow
from scipy import ndimage
```

```
for i in range(1,11) :
  loc = './train/cats/cats ({}).jpg'.format(i)
  img = cv.imread(loc)
  flipped = np.flipud(img) # 상하반전
  rotated = ndimage.rotate(img, 30) # 회전
  gauss = ndimage.gaussian_filter(img, sigma = 3) # 가우시안 필터링
  loc = './train/cats/cats ({}).jpg'.format(i+40)
  cv.imwrite(loc, rotated)
  loc = './train/cats/cats ({}).jpg'.format(i+60)
  cv.imwrite(loc, flipped)
  loc = './train/cats/cats ({}).jpg'.format(i+80)
  cv.imwrite(loc, gauss)
  
```

```
for i in range(1,10) :
  loc = './train/dogs/ddog ({}).jpg'.format(i)
  img = cv.imread(loc)
  flipped = np.flipud(img)
  rotated = ndimage.rotate(img, 30)
  gauss = ndimage.gaussian_filter(img, sigma = 3)
  loc = './train/dogs/ddog ({}).jpg'.format(i+36)
  cv.imwrite(loc, rotated)
  loc = './train/dogs/ddog ({}).jpg'.format(i+56)
  cv.imwrite(loc, flipped)
  loc = './train/dogs/ddog ({}).jpg'.format(i+76)
  cv.imwrite(loc, gauss)
```

```
import os
loc = './train/dogs'
imgList = os.listdir(loc) # 이 밑에 file 들이 list 로 넘어옴
imgList
```

```
# random 으로 사진 선택 dog
from matplotlib import pyplot as plt

index = np.random.choice(len(imgList)-20,5) # list len 중에 5 개 숫자 뽑음
for i in index :
  loc = './train/dogs/dog ({}).jpg'.format(i)
  print(loc)
  img = cv.imread(loc)
  if i < 37 :
    img = ndimage.rotate(img, 15)
    img = np.flipud(img)
  if i > 36 and i < 57 :
    img = ndimage.rotate(img, 15)
  if i > 56 and i < 77 :
    img = ndimage.gaussian_filter(img, sigma = 3)
  if i > 76 :
    img = np.flipud(img)
   
  plt.imshow(img)
  plt.show()
  loc = './train/dogs/dog ({}).jpg'.format(i+100)
  cv.imwrite(loc, img)
```

```
#cat

loc = './train/cats'
imgList = os.listdir(loc) # 이 밑에 file 들이 list 로 넘어옴
imgList

index = np.random.choice(len(imgList)-25,5) # list len 중에 5 개 숫자 뽑음

for i in index :
  loc = './train/cats/cat ({}).jpg'.format(i)
  print(loc)
  img = cv.imread(loc)
  if i < 41 :
    img = ndimage.rotate(img, 15)
    img = np.flipud(img)
  if i > 40 and i < 61 :
    img = ndimage.rotate(img, 15)
  if i > 60 and i < 81 :
    img = ndimage.gaussian_filter(img, sigma = 3)
  if i > 80 :
    img = np.flipud(img)
   
  plt.imshow(img)
  plt.show()
  loc = './train/cats/cat ({}).jpg'.format(i+200)
  cv.imwrite(loc, img)
```
